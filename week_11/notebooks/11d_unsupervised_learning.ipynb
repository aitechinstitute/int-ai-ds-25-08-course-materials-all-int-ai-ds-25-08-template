## 1. Theory: What is Unsupervised Learning?

### ðŸŽ¯ The Big Difference

**Supervised Learning:**
```
We have labels (answers)

Customer | Age | Income | Purchased
---------|-----|--------|----------
   A     | 25  | 50K    |   YES    â† We know this!

```

```

   B     | 30  | 60K    | ???  â† Unknown!

Task: Find hidden patterns in data
```

## 3. Method 1: K-Means Clustering

### ðŸ“– Theory: How K-Means Works

**Algorithm Steps:**
```
1. Choose K (number of clusters)
2. Randomly place K centroids
3. Assign each point to nearest centroid
4. Move centroids to center of assigned points
5. Repeat steps 3-4 until convergence
```

**Visual Example:**
```
Iteration 1:          Iteration 2:          Final:
   â—  â—                  â—                     â—
  â— â— â—     â†’         â—â—â—â—â—      â†’         â—â—â—â—â—
   â— â—                  â—                     â—
    X                    X                     X
(random)          (adjusting)           (converged)
X = centroid
```

**Key Idea:** Minimize within-cluster variance

---

### âš™ï¸ Key Parameters

**n_clusters (K):**
```
How many groups?
Must choose before running!
```

**random_state:**
```
For reproducibility
Same results every run
```

---

### ðŸŽ¯ Practice: Apply K-Means

````python
# Prepare features
X = df[['Annual_Income_k', 'Spending_Score']].values

print(f"âœ… Feature matrix shape: {X.shape}")
````

````python
# Scale features (important for distance-based algorithms!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"âœ… Data scaled (mean=0, std=1)")
````

````python
# Create K-Means model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

print("âœ… K-Means model created (K=4)")
````

````python
# Fit model and get cluster labels
df['KMeans_Cluster'] = kmeans.fit_predict(X_scaled)

print("âœ… Clustering complete!")
````

## 4. Choosing K: The Elbow Method

### â“ How Many Clusters?

**The Problem:**
```
K-Means needs K specified upfront
But how do we choose K?
```

**Solution: Elbow Method**
```
1. Try different values of K (e.g., 2 to 10)
2. Calculate "inertia" (within-cluster sum of squares)
3. Plot K vs Inertia
4. Look for "elbow" - where adding clusters stops helping much
```

**What is Inertia?**
```
Sum of squared distances from points to their cluster center
Lower = Better (tighter clusters)

But more clusters always â†’ lower inertia
Need balance!
```

---

````python
# Try K from 2 to 10
inertias = []
K_range = range(2, 11)

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(X_scaled)
    inertias.append(kmeans_temp.inertia_)

print("âœ… Tested K from 2 to 10")
````

````python
# Plot elbow curve
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=list(K_range),
    y=inertias,
    mode='lines+markers',
    marker=dict(size=10, color='blue'),
    line=dict(width=3)
))

fig.update_layout(
    title='Elbow Method: Finding Optimal K',
    xaxis_title='Number of Clusters (K)',
    yaxis_title='Inertia (Within-Cluster Sum of Squares)',
    template='plotly_white',
    width=800,
    height=500,
    showlegend=False
)

fig.show()
````

ðŸ’¡ **Look for the "elbow" - where the curve bends sharply!**

After that point, adding more clusters doesn't improve much.

## 5. Evaluation Metrics

### ðŸ“ How Good Are Our Clusters?

**Silhouette Score:**
```
Measures how similar a point is to its own cluster
vs other clusters

Range: -1 to +1
+1: Perfect (point very close to own cluster)
 0: Overlapping clusters
-1: Wrong cluster assignment
```

**Davies-Bouldin Score:**
```
Measures average similarity between clusters

Lower = Better
0 = Perfect separation
```

---

````python
# Calculate Silhouette Score
silhouette = silhouette_score(X_scaled, df['KMeans_Cluster'])

print(f"Silhouette Score: {silhouette:.3f}")
````

````python
# Calculate Davies-Bouldin Score
davies_bouldin = davies_bouldin_score(X_scaled, df['KMeans_Cluster'])

print(f"Davies-Bouldin Score: {davies_bouldin:.3f} (lower is better)")
````

## 6. Method 2: Hierarchical Clustering

### ðŸ“– Theory: How It Works

**Agglomerative (Bottom-Up) Approach:**
```
Step 1: Each point is its own cluster
   â—  â—  â—  â—  â—

Step 2: Merge closest pairs
  (â— â—) â—  â—  â—

Step 3: Keep merging
  (â— â—)(â— â—) â—

Step 4: Until one cluster
  (((â— â—)(â— â—)) â—)
```

**Key Advantage:** Don't need to specify K upfront!

**Dendrogram:** Tree diagram showing merge history

---

### ðŸ”— Linkage Methods

**How to measure distance between clusters?**

**Single Linkage:**
```
Distance = Minimum distance between any two points
Creates long chains
```

**Complete Linkage:**
```
Distance = Maximum distance between any two points
Creates compact clusters
```

**Average Linkage:**
```
Distance = Average of all pairwise distances
Balanced approach
```

**Ward:**
```
Minimizes within-cluster variance
Most popular for general use
```

---

### ðŸŽ¯ Practice: Apply Hierarchical Clustering

````python
# Create hierarchical clustering model
hierarchical = AgglomerativeClustering(n_clusters=4, linkage='ward')

print("âœ… Hierarchical model created (Ward linkage)")
````

````python
# Fit and predict
df['Hierarchical_Cluster'] = hierarchical.fit_predict(X_scaled)

print("âœ… Hierarchical clustering complete")
````

````python
# Visualize results
fig = px.scatter(
    df,
    x='Annual_Income_k',
    y='Spending_Score',
    color='Hierarchical_Cluster',
    title='Hierarchical Clustering Results',
    labels={'Annual_Income_k': 'Annual Income ($k)', 
            'Spending_Score': 'Spending Score',
            'Hierarchical_Cluster': 'Cluster'},
    template='plotly_white',
    width=900,
    height=600,
    color_continuous_scale='plasma'
)

fig.update_traces(marker=dict(size=10, opacity=0.7))
fig.show()
````

### ðŸŒ³ Dendrogram Visualization

**Shows the merge hierarchy!**

````python
# Create linkage matrix for dendrogram
# Use first 50 samples for clarity
sample_indices = np.random.choice(len(X_scaled), 50, replace=False)
X_sample = X_scaled[sample_indices]

linkage_matrix = linkage(X_sample, method='ward')

print("âœ… Linkage matrix computed")
````

````python
# Plot dendrogram
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.title('Hierarchical Clustering Dendrogram (50 samples)', fontsize=14)
plt.xlabel('Customer Index', fontsize=12)
plt.ylabel('Distance', fontsize=12)

dendrogram(linkage_matrix)
plt.axhline(y=10, color='r', linestyle='--', label='Cut for 4 clusters')
plt.legend()
plt.tight_layout()
plt.show()
````

ðŸ’¡ **How to read the dendrogram:**
- Height of merge = distance between clusters
- Horizontal line cuts determine number of clusters
- Higher merge = more different clusters

## 7. Method 3: DBSCAN

### ðŸ“– Theory: Density-Based Clustering

**How DBSCAN Works:**
```
Finds clusters based on density of points

Dense regions = Clusters
Sparse regions = Noise/Outliers
```

**Key Concepts:**

**1. Epsilon (Îµ):**
```
Radius to search for neighbors
Larger Îµ â†’ Fewer, larger clusters
```

**2. Min Samples:**
```
Minimum points needed to form dense region
Higher value â†’ Stricter clusters
```

**3. Point Types:**
```
Core Point: Has >= min_samples neighbors
Border Point: Near core point, but < min_samples neighbors  
Noise Point: Not in any cluster
```

---

### ðŸ’ª Advantages of DBSCAN

âœ… **Don't need to specify K**

âœ… **Finds arbitrary shapes** (not just circles)

âœ… **Identifies outliers** (noise points)

âœ… **Robust to outliers**

---

### ðŸŽ¯ Practice: Apply DBSCAN

````python
# Create DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)

print("âœ… DBSCAN model created (eps=0.5, min_samples=5)")
````

````python
# Fit and predict
df['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)

print("âœ… DBSCAN clustering complete")
````

````python
# Check number of clusters found
n_clusters = len(set(df['DBSCAN_Cluster'])) - (1 if -1 in df['DBSCAN_Cluster'] else 0)
n_noise = list(df['DBSCAN_Cluster']).count(-1)

print(f"Clusters found: {n_clusters}")
print(f"Noise points: {n_noise}")
````

````python
# Visualize DBSCAN results
# Note: -1 label means noise/outliers
fig = px.scatter(
    df,
    x='Annual_Income_k',
    y='Spending_Score',
    color='DBSCAN_Cluster',
    title='DBSCAN Clustering (Density-Based)',
    labels={'Annual_Income_k': 'Annual Income ($k)', 
            'Spending_Score': 'Spending Score',
            'DBSCAN_Cluster': 'Cluster (-1 = Noise)'},
    template='plotly_white',
    width=900,
    height=600,
    color_continuous_scale='jet'
)

fig.update_traces(marker=dict(size=10, opacity=0.7))
fig.show()
````

ðŸ’¡ **Notice:** DBSCAN automatically found clusters and outliers!

Points marked as -1 are considered noise/outliers.

## 8. Compare All Methods

### ðŸ“Š Side-by-Side Comparison

````python
# Create comparison visualization
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=('K-Means', 'Hierarchical', 'DBSCAN'),
    horizontal_spacing=0.1
)

# K-Means
fig.add_trace(
    go.Scatter(
        x=df['Annual_Income_k'],
        y=df['Spending_Score'],
        mode='markers',
        marker=dict(color=df['KMeans_Cluster'], size=8, colorscale='viridis'),
        showlegend=False
    ),
    row=1, col=1
)

# Hierarchical
fig.add_trace(
    go.Scatter(
        x=df['Annual_Income_k'],
        y=df['Spending_Score'],
        mode='markers',
        marker=dict(color=df['Hierarchical_Cluster'], size=8, colorscale='plasma'),
        showlegend=False
    ),
    row=1, col=2
)

# DBSCAN
fig.add_trace(
    go.Scatter(
        x=df['Annual_Income_k'],
        y=df['Spending_Score'],
        mode='markers',
        marker=dict(color=df['DBSCAN_Cluster'], size=8, colorscale='jet'),
        showlegend=False
    ),
    row=1, col=3
)

fig.update_xaxes(title_text="Income ($k)", row=1, col=1)
fig.update_xaxes(title_text="Income ($k)", row=1, col=2)
fig.update_xaxes(title_text="Income ($k)", row=1, col=3)

fig.update_yaxes(title_text="Spending Score", row=1, col=1)

fig.update_layout(
    title_text="Clustering Method Comparison",
    height=500,
    width=1400,
    template='plotly_white'
)

fig.show()
````

### ðŸ“ˆ Evaluation Metrics Comparison

````python
# Calculate metrics for each method
methods = ['K-Means', 'Hierarchical', 'DBSCAN']
cluster_columns = ['KMeans_Cluster', 'Hierarchical_Cluster', 'DBSCAN_Cluster']

results = []

for method, col in zip(methods, cluster_columns):
    # Skip noise points for DBSCAN
    if method == 'DBSCAN':
        mask = df[col] != -1
        X_eval = X_scaled[mask]
        labels = df[col][mask]
    else:
        X_eval = X_scaled
        labels = df[col]
    
    silhouette = silhouette_score(X_eval, labels)
    davies = davies_bouldin_score(X_eval, labels)
    n_clusters = len(set(labels))
    
    results.append({
        'Method': method,
        'Clusters': n_clusters,
        'Silhouette': silhouette,
        'Davies-Bouldin': davies
    })

comparison_df = pd.DataFrame(results)
comparison_df
````

## 9. Choosing the Right Method

### ðŸŽ¯ Decision Guide

**Use K-Means when:**
```
âœ… Know number of clusters
âœ… Clusters are spherical
âœ… Similar sized clusters
âœ… Need fast algorithm
âœ… Large dataset
```

**Use Hierarchical when:**
```
âœ… Don't know K upfront
âœ… Want to see hierarchy
âœ… Small to medium dataset
âœ… Need dendrogram
âœ… Clusters within clusters
```

**Use DBSCAN when:**
```
âœ… Arbitrary shaped clusters
âœ… Have outliers to identify
âœ… Unknown number of clusters
âœ… Varying cluster densities
âœ… Geographic data
```

---

### ðŸ“‹ Quick Comparison

| Feature | K-Means | Hierarchical | DBSCAN |
|---------|---------|--------------|--------|
| **Need to specify K** | Yes | Yes | No |
| **Cluster shape** | Spherical | Any | Any |
| **Outlier detection** | No | No | Yes |
| **Speed** | Fast | Slow | Medium |
| **Scalability** | High | Low | Medium |
| **Deterministic** | No* | Yes | Yes |

*Depends on random initialization
